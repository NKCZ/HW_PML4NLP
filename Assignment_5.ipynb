#### 5. Make a Concordance

a. Create a concordance display for an arbitrary word. See the example below  

![](../../Data/figs/Sample-concordance-lines-of-actually.png)

```python
# a)

def Concordance(text:str, keyword:str = "the") -> str:
    import string
    keyword = keyword.lower()
    text_lower = text.lower()
    text_list = text_lower.split()
    #text_list = [re.sub(r"[^a-z]", "", x) for x in text_list]# just for the sake of being consistent with not using re
    text_list = [w.strip(string.punctuation) for w in text_lower.split()]
    text_list = [x for x in text_list if x!=""]

    #concordance
    indices = [i for i, x in enumerate(text_list) if x == keyword]

    s="#\t\tPosition-5\t\tkeyword\t\tPosition+5\n"

    if len(indices)>10:
        for i in range(10):
            s+=f"{i+1}\t\t{' '.join(text_list[indices[i]-5:indices[i]])}\t\t{text_list[indices[i]]}\t{' '.join(text_list[indices[i]+1: indices[i]+6])}\n"
    else:
        for i in range(len(indices)):
            s+=f"{i+1}\t\t{' '.join(text_list[indices[i]-5:indices[i]])}\t\t{text_list[indices[i]]}\t{' '.join(text_list[indices[i]+1: indices[i]+6])}\n"
    print(s)
    return s

programming = Concordance(text, "programming")
```
